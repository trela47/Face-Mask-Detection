```python
import cv2
import matplotlib.pyplot as plt


videocap = cv2.VideoCapture(0) #uses the webcam for video input
# videocap = cv2.VideoCapture('file_name.mp4')    #option for entering video file as an input


onceuponatime = 0
while True and onceuponatime < 25:
    _, img = videocap.read()    #retuns a flag and an image ('frame')        
    onceuponatime+=1
videocap.release()    #stops the video


img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(img2, 1.1, 4)    #used to detect where the face is

n=0
#this will frame the face in a rectangle
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
    n+=1


img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```
